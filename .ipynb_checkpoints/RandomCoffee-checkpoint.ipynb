```python
def pairSignUps(n):
    if n % 2 != 0:
        return []
    participantIDs = list(range(n))
    np.random.shuffle(participantIDs)
    pairs = []
    while len(participantIDs) > 0:
        first = participantIDs.pop()
        second = participantIDs.pop()
        pairs.append([first,second])
    return pairs

# print(pairSignUps(10))
# print(pairSignUps(9))  
```
